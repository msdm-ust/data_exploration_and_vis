```python
# Filter the transaction data by the currency column with the value GBP
is_GBP = transactions_data['CURRENCY']=='GBP'
# drop out the rows in the transaction data which is not GBP currency
GBP_transactions =transactions_data[is_GBP]

GBP_transactions.head()
```

```python
# Filter the transaction data to exclude the transactions not in USD
not_USD = transactions_data['CURRENCY']!='USD'
not_USD_trans = transactions_data[not_USD]
not_USD_trans.head(4)
not_USD_trans.describe()
# the output shows 657109 transactions not in USD.
```

```python
# extract the USD data 
is_USD = transactions_data['CURRENCY']=='USD'
USD_trans = transactions_data[is_USD]
USD_trans.head()
print('the average transactions in USD is ',np.mean(USD_trans['AMOUNT']))
print('the median transaction in USD is ', np.median(USD_trans['AMOUNT']))
```

```python
currencys = pd.Series.unique(transactions_data['CURRENCY'])
currencys
```

```python
#  function to output the number of particular currency transactions
# store all currency transactions in a list
def num_currency(array):
    transactions = []
    for i in array:
        trans = transactions_data[transactions_data['CURRENCY']==i]
        transactions.append(len(trans)+1)
    return transactions
transactions=num_currency(currencys)
data = {'CURRENCY':currencys,'Number of Transactions':transactions}
df = pd.DataFrame(data)
df
```
